```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read all the csv file with student marks
sem1=pd.read_csv("/content/Sem_1.csv")
sem2=pd.read_csv("/content/Sem_2.csv")
sem3=pd.read_csv("/content/Sem_3.csv")
sem3.head()
```

```python
from pandas.core.internals.blocks import final
final_result= sem1.merge(sem2,on='Student_id').merge(sem3,on='Student_id')
final_result["Physics"]=final_result["Physics_x"]+final_result["Physics_y"]+final_result["Physics"]
final_result["Chemistry"]=final_result["Chemistry_x"]+final_result["Chemistry_y"]+final_result["Chemistry"]
final_result["Math"]=final_result["Math_x"]+final_result["Math_y"]+final_result["Math"]
final_result["Biology"]=final_result["Biology_x"]+final_result["Biology_y"]+final_result["Biology"]
final_result.drop(columns=["Physics_x","Math_x","Chemistry_x","Biology_x",	"Physics_y","Math_y",	"Chemistry_y","Biology_y"],inplace=True)

```

```python
final_result
```

```python
final_result
columns=list(final_result.columns[1:])
for i in columns:
  final_result[i]=['Fail' if ((x/300)*100)<30 else 'Pass' for x in final_result[i]]
```

```python
final_result
```

```python
passed_physics,failed_physics=final_result['Physics'].value_counts()
passed_chemistry,failed_chemistry=final_result['Chemistry'].value_counts()
passed_biology,failed_biology=final_result['Biology'].value_counts()
passed_maths,failed_maths=final_result['Math'].value_counts()

sub=["Physics","Chemistry","Biology","Math"]
data_pass=[passed_physics,passed_chemistry,passed_biology,passed_maths]
data_fail=[failed_physics,failed_chemistry,failed_biology,failed_maths]




# Numbers of pairs of bars you want
N = len(sub)


# Position of bars on x-axis
ind = np.arange(N)
# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, data_pass , width, label='Pass')
plt.bar(ind + width, data_fail, width, label='Fail')

plt.xlabel('All semester')
plt.ylabel('No of Pass and Fail in each subject')
plt.title('Plot')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, sub)

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```

```python
sem1_p=sem1['Physics'].value_counts()['Pass']
sem2_p=sem2['Physics'].value_counts()['Pass']
sem3_p=sem3['Physics'].value_counts()['Pass']
data_physics=[sem1_p,sem2_p,sem3_p]
sem1_c=sem1['Chemistry'].value_counts()['Pass']
sem2_c=sem2['Chemistry'].value_counts()['Pass']
sem3_c=sem3['Chemistry'].value_counts()['Pass']
data_chemistry=[sem1_c,sem2_c,sem3_c]
sem1_b=sem1['Biology'].value_counts()['Pass']
sem2_b=sem2['Biology'].value_counts()['Pass']
sem3_b=sem3['Biology'].value_counts()['Pass']
data_biology=[sem1_b,sem2_b,sem3_b]
sem1_m=sem1['Math'].value_counts()['Pass']
sem2_m=sem2['Math'].value_counts()['Pass']
sem3_m=sem3['Math'].value_counts()['Pass']
data_math=[sem1_m,sem2_m,sem3_m]
```

```python
sem=["Semester_1","Semester_2","Semester_3"]
N=len(sem)
ind = np.arange(N)

plt.figure(figsize=(10,5)) 
width = 0.1


plt.bar(ind, data_physics , width, label='Physic')
plt.bar(ind + width, data_chemistry, width, label='Chemistry')
plt.bar(ind + 2*width, data_biology, width, label='Biology')
plt.bar(ind + 3*width, data_math, width, label='Math')



plt.xlabel('All semester result')
plt.ylabel('No of Pass in each subject')
plt.title('Plot')
plt.xticks(ind, sem)
plt.legend(loc='best')
plt.show()


      
```
